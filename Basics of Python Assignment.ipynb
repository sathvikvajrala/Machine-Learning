1. Prime Numbers

```python
n = int(input("Enter a number: "))
c = 0
for i in range(2, int(n/2)+1): #number in range from 2 to half of the n
    if n%i==0:
        c+=1
    break
if c==0 and n > 1:
    print(n, "is a prime number")
else:
    print(n, "is not a prime number")
```

5. Palindrome

```python
def is_palindrome(s):
    if s==s[::-1]:   # reverses the string
        return True
    else:
        return False
s=input("Enter string: ")
is_palindrome(s)
```
